```python
from sklearn.decomposition import PCA
from node2vec import Node2Vec as n2v

WINDOW = 10 # Node2Vec fit window
MIN_COUNT = 1 # Node2Vec min. count
BATCH_WORDS = 4 # Node2Vec batch words

g_emb = n2v(
  G,
  dimensions=20,
  walk_length=16,
  num_walks=100
)

mdl = g_emb.fit(
    vector_size = 16,
    window=WINDOW,
    min_count=MIN_COUNT,
    batch_words=BATCH_WORDS
)

input_node = '370913'
for s in mdl.wv.most_similar(input_node, topn = 10):
    print(s)
```

```python
emb_df = (
    pd.DataFrame(
        [mdl.wv.get_vector(str(n)) for n in G.nodes()],
        index = G.nodes
    )
)
```

```python
emb_df
```

```python
pca = PCA(n_components = 2, random_state = 7)
pca_mdl = pca.fit_transform(emb_df)

emb_df_PCA = (
    pd.DataFrame(
        pca_mdl,
        columns=['x','y'],
        index = emb_df.index
    )
)
plt.clf()
fig = plt.figure(figsize=(6,4))
plt.scatter(
    x = emb_df_PCA['x'],
    y = emb_df_PCA['y'],
    alpha = 0.5
)
plt.xlabel('PCA-1')
plt.ylabel('PCA-2')
plt.title('PCA Visualization')
plt.plot()
```

```python
from sklearn.model_selection import train_test_split
SEED = 5

X= modified_df.drop(["avg_rating"],axis =1)
y= modified_df["avg_rating"]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=SEED)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca.fit(X_train)
```

```python
variance = pca.explained_variance_ratio_
```

```python
import numpy as np

var=np.cumsum(np.round(variance, decimals=3)*100)
x_axis = []
for i in range(6):
  x_axis.append(i+1)
print(x_axis)
print(var)
```

```python
import seaborn as sns

plt.ylabel('% Variance Explained')
plt.xlabel('# of Features')
plt.title('PCA Analysis')
# plt.ylim(30,100.5)
plt.style.context('seaborn-whitegrid')
sns.despine(top=True, right=True, left=False, bottom=False)
plt.plot(var, color="0.5")
np.interp(40, x_axis, var)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

pipeline_linear_regression =Pipeline([("s1", StandardScaler()),
                      ("pca1", PCA(n_components=1)),
                      ("Linear Regression", LinearRegression())])

pipeline_knn = Pipeline([("s2", StandardScaler()),
                       ("pca2",PCA(n_components=4)),
                       ("KNN",KNeighborsRegressor())])
```

```python
from sklearn.model_selection import cross_val_score

pipelines = [pipeline_linear_regression, pipeline_knn]

pipeline_dict = {0: "LinearRegression", 1: "KNeighbors"}

for pipe in pipelines:
    pipe.fit(X_train, y_train)

cv_results_rms = []
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train,y_train,scoring="neg_root_mean_squared_error", cv=10)
    cv_results_rms.append(cv_score)
    print("%s: %f " % (pipeline_dict[i], cv_score.mean()))
```

```python
import sklearn.metrics as metrics

# Model prediction on test data
pred = pipeline_linear_regression.predict(X_test)

# Model Evaluation
r2 = metrics.r2_score(y_test, pred)
adjusted_r2 = 1 - (1-metrics.r2_score(y_test, pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
mae = metrics.mean_absolute_error(y_test, pred)
mse = metrics.mean_squared_error(y_test, pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test, pred))

# initialise data of lists.
result = [[r2],[adjusted_r2],[mae],[mse],[rmse]]
# Creates pandas DataFrame.
scores = pd.DataFrame(result,columns= ["Scores"] ,index = ["R-Squared","Adjusted R-Squared", "Mean Absolute Error","Mean Square Error","Root Mean Square Error"])
scores
```
